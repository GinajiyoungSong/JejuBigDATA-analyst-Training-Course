```python
import numpy as np
F1 = np.array([[1,2],[3,4]]) # 2x2
F2 = np.array([[2,2],[2,2]]) # 2x2

print(F1)
print(F2)
print()

print( F1 + F2 ) # 요소끼리의 덧셈

print()
print( F1 * F2 )  #요소끼리의 곱셈

print()
print( F1 @ F2 ) # a@a 내적 = 행렬곱은 이것으로 함

# 1,2   @  2,2   => 6,6
# 3,4      2,2     14,14    # 두 벡터의 사이각을 구해 차원 확장

# 딥러닝 -> 한번에 여러개의 변수에서 2개로 축소시키는게 아니라 
#           조금씩 차원 축소( 차원축소는 즉 행렬곱: 내적 ) 해서 값을 얻어간다.
```

# 행렬은 나눗셈이 없어서 역행렬 곱해야 한다

# 역행렬  정의

###       A       x       = b
### (n x n ) (n x 1) = n x 1

# A 는  행렬(매트릭스) 



#  A^-1A=I가 되는 행렬

####    항등원인 숫자 1처럼 곱했을 때 

# AI=A, IA=A와 같이 자기자신이 되게 하는 행렬



#    역행렬=> I를 identity matrix
        
#   Ax = b → 양쪽에 역행렬 곱하면

#                                    ( A^-1Ax = A^-1b ) 

#                                 → ( I* x = A^-1b )→ x = A^-1b


```python
# 방정식 해 (x구하기)
A = np.array([[2,1,-2],[3,0,1],[1,1,-1]])
print(A)

print()
b = np.transpose(np.array([[-3,5,-2]]))  #열로 전치
print(b)

# A x = b
# x 벡터를 구해라 
```

# A x = b

# x 벡터를 구해라 

```python
# 방정식의 해


#        A 행렬 (3x3)     x    x라는 벡터(열벡터)                 =   b (열벡터) b값을 구해라
#  A의 행렬의 1열 수  (2) x (x1,1) +  (1) x (x,1,2) +  (-2) x (x,1,3) =    -3
#  A의 행렬의 2열 수  (3) x (x1,1) +  (0) x (x1,2)  +  ( 2) x (x1,3)  =    5
#  A의 행렬의 3열 수  (-1) x (x1,1) + (1) x (x1,2)  +  (-1) x (X1,3)  =   -2
```

# x 열벡터를 구하기 =>  Ax = b    => x = A^-1 * b

```python

x = np.linalg.solve(A,b)  #A 의 역행렬을 구해서 b와 곱해줌
print(x)
```

```python
np.linalg.inv(A) # A의 역행렬 구하기 # 행렬 계산기로 A 의 역함수를 구하면 이값이 나옴
```

```python
np.dot(np.linalg.inv(A),b)   #A의 역행렬에 b 곱해주면  행렬곱
```

```python
np.linalg.inv(A)@b   # 내적 @ 행렬에서는 곱셈이 내적이므로 => * 일반 곱셈은 그냥 요소끼리 곱셈
```

# 정방향행렬 (2x2) 행/렬 같은 * 역행렬 = identity( 단위행렬 )

```python
x = np.array([[1,2], [3,4]])
print(x)
print()

y = np.linalg.inv(x) # 역행렬함수 이용
print(y)


np.dot(x,y) # 행렬곱 = x 와 x의 역행렬 x^-1 둘을 곱하면 => 단위행렬 답이 나옴 (1 0)
            #                                                                 (0 1)
```

```python
c = np.array([[1, 1 ,1],[0, 2, 5],[2, 5, -1]])
print(c)

print()
d = np.transpose(np.array([[6,-4, 27]]))

# c * x = d 일때
x = np.linalg.solve(c,d) # x는 역행렬 
print(x)

print() # 다른 2가지 방법으로도 답
#역행렬 곱해주면 계산되니까
print(np.dot(np.linalg.inv(c),d))
np.linalg.inv(c)@d
```
