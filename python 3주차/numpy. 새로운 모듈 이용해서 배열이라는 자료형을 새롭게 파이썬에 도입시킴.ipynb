# 거듭제곱(행열 곱) : @ = > 너는 내적

```python
x = np.arange(6).reshape((3,2))
print(x)


x@x
```

```python
x@x => 에러 : 앞의 행렬 열수와 뒤의 행렬의 행수가 일치하지 않아서 에러
    # (3x2)(3x2) : 앞의 행렬 열수와 뒤의 행렬의 행수가 일치하지않아서 에러가 난다
    #행렬의 곱셈은 ( m x k ) @ ( k x n ) = ( m x n ) 이 되는 것
    
    위의 연산을 하려면 x 를 전치행렬시켜줘야함 => (2 x 3) 으로 
```

# Zip 으로 데이터 각각 요소별 묶기

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
print(xarr);print()


yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
print(yarr);print()


cond = np.array([True, False, True, False, True])
cond
```

```python
print(zip(xarr, yarr, cond))   
# zip : 요소별로 데이터 묶음, zip은 고유포멧이라서 안보이고
print()
                              # 그래서 list해서 보내줘야한다


print(list(zip(xarr, yarr, cond)))  

# 결과값 보면 1행 1열끼리, 2행 2열끼리 , 3행 3열끼리 계속 묶임               
```

```python
print()
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]   
# 3항 연산자
print("result = ",result)
```

```python
#True 일때는 xarr 에서 불러오고 False 일때는 yarr  => 결국 위에 조건문과 동일

re = np.where(cond, xarr, yarr) 
re
```

# 행렬을 나눗셈할때 => 역행렬을 구해서 곱한다

```python
       A x       = b
(n x n ) (n x 1) = n x 1

A 는  행렬(매트릭스) x 와 y n콜룸을 가진 벡터.

 A^-1A=I가 되는 행렬
    숫자 1처럼 곱했을 때 AI=A, IA=A와 같이 자기자신이 되게 하는 행렬
    역행렬=> I를 identity matrix
        
   Ax = b → 양쪽에 역행렬 곱하면 ( A^-1Ax = A^-1b ) 
                                 → ( I* x = A^-1b )→ x = A^-1b

    
    행렬은 나눗셈이 없어서 역행렬 곱해야 한다
```

# 수학적으로 참고할것

```python

- 정방행렬 : 정사각형(행과 열수가 같은 행렬)
- 대칭행렬 : 우상단과 좌하단이 같은 것
           : 모든 상관분석을 하기 위해 행과 열의 같은 변수를 준다. 
            
- 대각행렬 : scale을 변화시키는 요소 (대각행렬이 전부 1이면 다른거랑곱했을때 자신이 나옴)
- 단위행렬(identity) : 행렬곱의 항등원, 행렬과 역행렬을 곱하면 단위행렬

- 역행렬 : 행렬의 나눗셈을 구해서 곱해지는 행렬( 정방행렬만 가능 )
- 특이행렬 : 역행렬이 없는 행렬
    
- 직교행렬 : 축의 정직교
           : 각 축간의 내적이 0 인 행렬
            
-  행렬 분해 :행렬을 특정한 구조를 가지니 다른 행렬의 곱으로 나타내는 것
    - 3x3 행렬은 3x1 크기의 두행렬과 1x1 크기의 대각 행렬을 통해 나타낼수 있다.
    -선형 방정식의 해를 구하거나, 행렬 계산을 효율적으로 하거나, 행렬의 특정구조를 밝히는 목적
    -역행렬을 구하기 위해서


      
            
- 정방행렬이면서 대칭행렬 : 공분산행렬, 상관계수 행렬, 거리행렬
    
    
- 공분산 : 두 수간의 관계 ((한수 - 평균) * (다른수 - 평균)) / (n-1)
     2개의 확률변수의 상관정도를 나타내는 값
        만약 2개의 변수중 하나의 값이 상승하는 경향을 보일 때, 
        다른 값도 상승하는 경향의 상관관계에 있다면, 
        공분산의 값은 양수가 될 것이다. 
        반대로 2개의 변수중 하나의 값이 상승하는 경향을 보일 때, 
        다른 값이 하강하는 경향을 보인다면 공분산의 값은 음수가 된다. 

        
- 상관계수(표준화 -1 ~1) : 공분산 / 표준편차의 곱
```
