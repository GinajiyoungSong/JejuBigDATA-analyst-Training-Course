# 미분 : 

# 분산 : (sigma(관측치 - 평균)^2) / n-1

# 표준편차 : root(sigma(관측치 - 평균)^2) / n-1)

# 공분산 : sigma( x-xbar)(y-ybar)/n-1

# 상관계수 : (sigma( x-xbar)(y-ybar)/n-1) / (xsigma * ysigma) 

: -1 ~ 1 사이의 값을 가짐

# 벡터간 거리값 :  각 요소의 차의 제곱을 root 로 구함

# model 평가 : MSE(mean square errot ) : 연속 수치인 경우평가

#                      RMSE(root mean square errot)

# 신경망에서 cost function 비용함수 : ( 예측치 - 관측치 )

##                      => back propagation (역전파를 통해서 가중치를 수정)

```python
# 국어 수학점수 vector : 상관계수가 1이다. 양의 상관관계 => 국어점수가 높으면 수학점수도 높다
# 내적이 1이면 차이가 없다 : 즉 상관계수와 내적이 같음
# 상관계수를 구하는 식은 어려우니까 
@ 내적을 이용해서 해석
```

```python
# 회귀분석  상관분석

- 회귀분석에서 다중공선성 : 독립변수간의 영향력이 있으면 종속변수가 과하게 영향을 받을수 있음

    - 회귀분석에서 자기상관성(Autocorelation)이 존재 : => 시계열분석
        
        자기 상관성 (AR 모형 )
        어떤한 변수에 대해서 이전의 값이 이후의 값에 영향을 미치고 있는 상황을 말함
 


대표적인 시계열 모델 ( AR, MA, ARMA, ARIMA)

(MA 모델 - 이동 평균 : 시간이 지날수록 평균값이 지속적으로 감소하거나 증가할때 = 평균이동있음)

- 시계열 분석 ARMA (Autocorelation + Moving Average )
    회귀분석을 알아야만 시계열 분석 알수 있음 = > 시간을 고려한다는 점이 가장 큰 차이
    
   - 이동평균법 : 주파수를 줄여줌 = > 복잡하면 분석 불가
   - 정상성을 띈 데이터에만 가능

- 비정상 -> 정상성

-ARIMA 모형 (integrated 통합 => 비정상성을 띈 데이터에도 가능)



판다스 pandas => 시계열 분석을 지원 ( 시간 index 지원 )
```

```python
import numpy as np
from statistics import *
x=np.array([7,8,9])
y=np.array([9,10,20])        # 증가값이 큰차이를 보였으므로 증가하는 양상을 보이는 상관계수 나옴
print(y.mean());print()
# 중위수 => 이상치의 영향을 받지 않아서

# 공분산 행렬
print("공분산", np.cov(y))   # 공분산, 2개의 변수간의 상관정도 = 다변수 간의 상관관계
print()
#상관계수 행렬 (열 2, 행 2 : 4)
print("상관계수", np.corrcoef(x,y))
```

```python
points=np.arange(-5,5,0.01)
xs, ys = np.meshgrid(points, points)

# 중심 0.0 에서 부터 벡터 원중심에서 방향성가지고 퍼져나가는 거리값



z =np.sqrt(xs**2 + ys**2)
# square 행단위와 열단위로 각각 해당배열을 정방행렬로 선언
#    (행단위 xs) (열단위 ys)

# 높이값에 이미지 색값을 넣음

                             # 팔레트 : cmap = rainbow
plt.imshow(z, cmap=plt.cm.rainbow) # 이 레인보우값을 이미지로 만들어라
plt.colorbar()

plt.title('Image plot of $\sqrt{x^2+y^2}$ for a grid of values')   # : $ $ 사이에 수식넣기
plt.plot()
```
