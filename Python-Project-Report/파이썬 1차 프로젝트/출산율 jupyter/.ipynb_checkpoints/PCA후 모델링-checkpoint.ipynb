```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

data = pd.read_csv('team2.csv', index_col=0)
data
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

# MinMax

```python
# MinMax_Scale 
from sklearn.preprocessing import minmax_scale
minmax_X = minmax_scale(X)
minmax_Y = minmax_scale(y)

X_train, X_test, y_train, y_test = train_test_split(minmax_X,minmax_Y, random_state=0)
print(minmax_X.shape,minmax_Y.shape,'\n')

print(minmax_Y)
```

# PCA(16차원으로 축소)

```python

import numpy as np
from sklearn.decomposition import PCA


# 16개의 변수중 영향력있는 변수 3개추출
pca = PCA(n_components=0.95)
# pca = PCA(n_components=16)
model = pca.fit_transform(minmax_X)

print(pca.explained_variance_ratio_) # 설명력

# 설명력 누적시켜서 그래프로 그림
# elbow를 시각적으로 확인
plt.plot(np.cumsum(pca.explained_variance_ratio_)) 
plt.show()


plt.plot(pca.explained_variance_ratio_, 'bo-') 
# 분산이 계쏙줄어듬 = 첫번째 분산이 큰 C1 축에 직교하는 두번째 분산을 유지하는 C2
# 분산을 유지하는 C3... 계속 찾아가는 과정. 이 데이터에서는 4차원부터 매우 적음.
# 0.95 = 95%를 유지하는것을 찾아라
plt.show()
pca.explained_variance_ratio_
```

```python
# 변수 3개추출 후 DataFrame으로 생성
princ = pd.DataFrame(data=model[:,:4], columns=range(0,4))
princ
```

# PCA후 여러 모델에 적용

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler # kmean, PCA는 전처리에 민감하다.


X_train, X_test, y_train, y_test = train_test_split(princ,minmax_Y, random_state=0)

scaler=MinMaxScaler().fit(X_train) # 이렇게 연결하는 것을  chaining
X_train_scaled=scaler.transform(X_train)

svm=SVR()
svm.fit(X_train_scaled, y_train)

X_test_scaled=scaler.transform(X_test) # 테스트데이터도 해준다.

print('테스트 점수 : {:.2f}'.format(svm.score(X_test_scaled, y_test)*100), '%' )
```

# PCA 후에 테스트점수 87 % 로 상승함

# RandomForest Regreesor

```python
from sklearn.ensemble import RandomForestRegressor


X_train, X_test, y_train, y_test = train_test_split(princ,minmax_Y, random_state=0)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

X_pred=model.predict(X_train)
print('X_train MSE : ', mean_squared_error(y_train, X_pred))
y_pred = model.predict(X_test)
print("X_test MSE : ",mean_squared_error(y_test, y_pred))  # mse를 비교했을때 test>train : 과소적합 


df1 = pd.DataFrame({'y_test':y_test, 'y_preds':y_pred})
df1.corr()

# 오차가 너무 작아 이상함..
```

# XGboost

```python
import xgboost as xgb

X_train, X_test, y_train, y_test = train_test_split(princ,minmax_Y, random_state=0)

xg_reg = xgb.XGBRegressor()
xg_reg.fit(X_train,y_train)
preds = xg_reg.predict(X_test) 

#모델평가
rmse = np.sqrt(mean_squared_error(y_test,preds)) 
print("RMSE : %f" % (rmse))

xgb.plot_tree(xg_reg,num_trees=0)
# rc : resource control 한다

plt.rcParams['figure.figsize'] = [30,10]
plt.show()

X_pred=xg_reg.predict(X_train)
print('X_train MSE : ', mean_squared_error(y_train, X_pred))
y_pred = xg_reg.predict(X_test)
print("X_test MSE : ",mean_squared_error(y_test, y_pred))  # mse를 비교했을때 test>train : 과소적합 


df1 = pd.DataFrame({'y_test':y_test, 'y_preds':y_pred})
df1.corr()
```

# Linear = OLS 최소자승법 사용

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(princ,minmax_Y, random_state=0)

ols= sm.OLS(y_train, X_train)  
ols = ols.fit() # OLS 모델에 fitting 시킴

X_pred=ols.predict(X_train)
print('X_train MSE : ', mean_squared_error(y_train, X_pred))
y_pred = ols.predict(X_test)
print("X_test MSE : ",mean_squared_error(y_test, y_pred))  # mse를 비교했을때 test>train : 과소적합 


df1 = pd.DataFrame({'y_test':y_test, 'y_preds':y_pred})
print(df1.corr())

ols.summary()
```

## OLS Linear 이상함. 설명력 - 값.. 


## GridSearchCV 안됨

```python
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(princ,minmax_Y, random_state=0)

param_grid={'C':[0.001,0.01,0.1,1,10,100], 
           'gamma':[0.001,0.01,0.1,1,10,100]}

clf = GridSearchCV(ols, param_grid, cv = 5)
clf.fit(X_train, y_train)
print(clf.best_estimator_)
print(clf.best_params_)
```

```python
X_pred2 = clf.predict(X_train)
print('X_train MSE : ', mean_squared_error(y_train, X_pred2))
y_pred2 = clf.predict(X_test)
print("X_test MSE : ",mean_squared_error(y_test, y_pred2))

df2 = pd.DataFrame({'y_test':y_test, 'y_preds':y_pred2})
df2.corr()
```
