```python
revenue = int(input(" 수입 입력:"))
coats = int(input("지출 입력:"))
result = ""

if revenue > coast :
    result = '순수입: , abs(revenue-coast)
elif revenue == coast :
    result = '손익분기점'
else :
    result = '순손실:'', abs(revenue-coast)


print( "순수익 ", round(result, 2), "순손실", round(result,2) )
```

```python
import datetime
dt = datetime.datetime.now()   #함수는 () 사용해야함
print(dt)
print()

# 데이터입력
age=input("생년월인(19951031)을 입력하시오")

# 데이터 처리
year = int(age[0:4]) # 인덱싱 위치로 read    #마지막은 미포함
month = int(age[4:6])
day = int(age[6:])


# 빼기 연산
# 분산 : 상계방지 (제곱)
year_diff = dt.year - year
month_diff = dt.month - month
day_diff = dt.day -day
if day_diff < 0 :  # 일 내림 처리
    month_diff -=1
if month_diff <0 : # 월 내림 처리
    year_diff -=1
print("당신의 나이는", year_diff + 1)
if year_diff < 18 :
    print("당신은 미성년자입니다.")
else :
    print("당신은 미성년자가 아닙니다.")
```

```python

from random import randint
import sys

tot_game = 0   ######################총 게임수
win = 0       ###################### 이긴 경우수

while True:
    n = randint(1, 101) 
    tot_game +=1

    while True:
        number =input("사용자 입력 (Q to exit:")
        if number.upper() =="Q":
            print("총게임수", tot_game-1, "이긴 게임수", win) ########## 게임수 보여줘야지!!!!
            sys.exit(0) 
            
        inumber = int(number)
        if ( n == inumber):
            print("정답!")
            win+=1                       ############ 정답맞추면 이긴 게임수를 추가해야함
            break
        elif ( n > inumber ):
            print("높은 수를 선택하시요")
        else:
            print("작은 수를 선택하시요")   
```

#7) 총 진행횟수 하고
# quess횟수를 누적시킨 다음 
# 평균 시도 횟수를 출력하시요

```python



from random import randint
import sys

tot_game = 0
win = 0

tot_attempt=[] ##########전체 시도회수 추가 : 초기화

while True:
    n = randint(1, 101)
    tot_game +=1
    
    game=0     #######게임 0번째 스타트
    
# 사용자수 입력
    while True:
        number = input("사용자 입력 (Q to exit:")
        game +=1
        if number.upper() =="Q":
            print("총게임수", tot_game-1, "이긴 게임수", win)
            
            game_tot = sum(tot_attempt)                            #전체 시도횟수의 합
            print("평균시도횟수", game_tot/(tot_game -1))
            
            sys.exit(0) #프로그램 종료
        inumber = int(number)    
        if ( n == inumber):
            print("정답!")
            win+=1
            
            tot_attempt.append(game)
            
            break
        elif ( n > inumber ):
            print("높은 수를 선택하시요")
        else:
            print("작은 수를 선택하시요")   
```

```python
income = int(input("소득을 입력하시오:")) * 12
print("연봉:", income , end="만원 \n")

tax=0.0 #범위 지정해주면 round() 비슷?? # float형 초기화

if income<1000 :
    tax = income * 0.095
elif income<4000 :
    tax = 1000* 0.095 + (income - 1000)*0.019
elif income<8000 :
    tax = 1000*0.095 + 3000*0.19 + (income -4000)*0.28
else :
    tax = 1000*0.095 + 3000*0.19 + 4000*0.28 + (income - 8000)*0.37
    
print("총 세금은 " , round(tax,3), end="만원 입니다.")
    
```

```python
ele =int(input("초등학생 몇명:")) 
mid =int(input("청소년 몇명:")) 
adult =int(input("어른 몇명:")) 
old =int(input("경로우대 몇명:")) 

ele_tot = 5000*ele 
mid_tot = 10000*mid
adult_tot = 15000*adult 
old_tot = 3000*old


tot = ele_tot + mid_tot + adult_tot + old_tot

# if 문을 사용하지 않고 선택
dis = int(input("카드없음(0), 일반카드(1), VIP(2)를 선택해주세요"))
select = [ tot, tot*0.9, tot*0.8]
print("총 요금은", select[dis], "원 입니다.")
```
