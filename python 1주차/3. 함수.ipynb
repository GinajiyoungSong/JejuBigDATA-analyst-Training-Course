# 팩토리얼 함수로 만들기

```python
# 수학에서 factorial 5! = 5*4*3*2*1
def factorial(n):
    sum = 1                     #곱해줄때 0 이면 안돼  # 항등원
    for i in range(n, 1, -1):                          # 곱셈의 항등원은 1, 덧셈의 항등원 0, 행렬의 항등원 : 단위행렬
        sum*= i
    return sum
factorial(5)
```

# 재귀함수 : 함수가 자기자신을 호출

```python
def factorial(n, stop=1):
    if n <=stop or n <=0: # 종료조건 - 종료조건이 없으면 계속 step over flow
        return 1
    return n*factorial(n-1, stop) # 함수가 자기조건을 호출하려면 종료조건을 정해줘야함
factorial(5)                       # n=5 * factorial(n-1, stop) n-1 의 팩토리얼 함수 호출
                                   # 5 * n=4 * factorial(n-1, stop) n-1 의 팩토리얼 함수 호출
                                   # 5 * 4 * n =3 * factorial(n-1, stop) n-1 의 팩토리얼 함수 호출
        
                                   #                  n=1 일때 stop =1 과 같은값으로 종료조건에 n =stop 에 걸려서 return 1
```

# power 함수를 구현하시요 (2^10 ) # pow(2,10)

```python
def power(r,n):
    value = 1
    for i in range(1, n+1):
        value = r * value
    return value
print(power(2,3))
```

# generator

```python
# for in : Collection( list, dict, set, tuple), iterator(ranges) :어제까지 배움
#        : generator, eneumerate

import random

def gaus_dist(n):
    while n > 0:              # 무한루프 안에
        yield random.random() # yield : 비동기 호출함수 0~1
        n-=1
                                                                                             #  n-=1
gd = gaus_dist(5) # 함수 실행이 되서 5 데이터를 함수에 넣고 yield 에서 값 하나 호출 해내고 ( n=5 인 수가 4가 되고 ) 다시 함수 실행
for i in gd:      # random.random 에 의해서 수 하나 넘겨주고 다음 순서, 총 i는 gd 변수 = [ yield ] 값을 하나하나 가져옴
    print(i)
```

# clouser 클로저의 응용 ( decorator : 장식자)

```python
def format_and_pad(template, space):
    def formatter(x):
        return (template % x).rjust(space)
    return formatter

fmt=format_and_pad('%.4f', 15) # 클로저를 응용해서 바깥 정의함수에 template = %.4f    space = 15
print(fmt(1.756))    # %.4f float:실수
print(fmt(2.222))

fmt2=format_and_pad('%d', 15)  # %d 정수   # %s string :문자열
print(fmt2(1.756))
```

# 이자율이 주어졌을 때 기간거치후의 총액과 이자수익을 출력하시오

```python
# 복리계산 = (원가1 + 0.04) ** 거치년수
interest_rate = 0.04
# 매개변수 = 원금, 거치기간

def getInput(): #입력
    principal = eval(input("예금액 :"))
    numberOfYears = eval(input("기간 :"))
    return(principal, numberOfYears)

def balanceAndInterest(prin,numYears): #처리
    balance = prin * ((1 + interest_rate)**numYears) # 이자 포함 계산액
    interestEarned = balance - prin                 # 순수 이자액
    return( balance, interestEarned)
def displayOutput(bal, intEarned): #출력
    print("예금 총액 : {0:,.2f}원   이자 총액 : {1:,.2f}원".format(bal,intEarned))

def main():
    principal, numberOfYears = getInput()
    bal, intEarned = balanceAndInterest(principal, numberOfYears)
    displayOutput(bal, intEarned)

main()
```
