# 산술연산


```python
a = 3
b = 4

print( a + b, a - b, a * b, a / b )
print( a/b )
print( a % b ) # 나머지 연산자 : 경우의 수를 생성할 때 : 변수가 가질수 있는 경우수를 확인
print()
print( 1 % 2 ) # 2 로 나눴을때 0 나머지, 1나머지 : 2가지 경우수밖에 없음
print( 2 % 2 )
print( 3 % 2 )
print( 4 % 2 )
print()

print( 7 / 4 )
print( 7 // 4 ) #몫연산자
print( 7 % 4) # 나머지 연산자
```

# 증가변수, 누적변수 산술연산

```python

num1 = 6
num1 += 1 # num1 = num1 + 1
num2 = 7
num2 -=5 # num2 = num2 -1
num3 = 8
num3 /=2 # num3 = num3/2
print(num1, num2, round(num3))
num1 = 1
num1 *= 3
num2 = 2
num2 **=3 # ^, ** 승수
print(num1, num2)
```

# 연산으로는 부족해서 -> 내장함수 존재


```python


print("sum([1,2,3]):", sum([1,2,3]))
print("min([1,2,3]):", min([1,2,3]))
print("abs(-11):", abs(-11))
print("pow(2,10):", pow(2,10)) #power 2의 10승
print("round(3.141595):", round(3.141595)) #round 반올림
print("round(3.141595):", round(3.141595,2 )) # ,2 자리수까지 반올림

```
