```python
# 다음 표(color와 mode에 따른 날씨)를 if문을 이용하여 출력하시오
#        ( 반응에 따른 날씨 예측 프로그램 )
#                      BLUE              RED
#         STEADY       화창              비오는 날
#        FLASHING      구름              눈오는 날

color = input("color 입력 (BLUE or RED):")
mode = input(" mode 입력 (STEADY or FLASHING)")


color = color.upper() # 소문자로 쓸지 대문자로 쓸지 모르므로
mode = mode.upper()  # 소문자를 대문자로 바꿔주는 upper() 함수 이용
result=""
if color =='BLUE' :
    if mode =='STEADY': 
        result = '화창한 날' 
    else : 
        result = '구름낀 날'         
else : 
    if mode =='STEADY' :
        result = '비오는 날'
    else : 
        result = '눈오는 날'
print("예측된 날씨는", result)
```

#5) 무한루프로 진행하던 게임은 q가 눌리면 종료합니다

# 무한반복을 계속하게 만들었으니 
# 몇회만에 맞추었는지를 확인하여 출력해 줍니다.

```python

from random import randint
import sys

tot_game = 0   ######################총 게임수
win = 0       ###################### 이긴 경우수

while True:
    n = randint(1, 101) 
    tot_game +=1

    while True:
        number =input("사용자 입력 (Q to exit:")
        if number.upper() =="Q":
            print("총게임수", tot_game-1, "이긴 게임수", win) ########## 게임수 보여줘야지!!!!
            sys.exit(0) 
            
        inumber = int(number)
        if ( n == inumber):
            print("정답!")
            win+=1                       ############ 정답맞추면 이긴 게임수를 추가해야함
            break
        elif ( n > inumber ):
            print("높은 수를 선택하시요")
        else:
            print("작은 수를 선택하시요")   
```

#7) 총 진행횟수 하고
# quess횟수를 누적시킨 다음 
# 평균 시도 횟수를 출력하시요

```python



from random import randint
import sys

tot_game = 0
win = 0

tot_attempt=[] ##########전체 시도회수 추가 : 초기화

while True:
    n = randint(1, 101)
    tot_game +=1
    
    game=0     #######게임 0번째 스타트
    
# 사용자수 입력
    while True:
        number = input("사용자 입력 (Q to exit:")
        game +=1
        if number.upper() =="Q":
            print("총게임수", tot_game-1, "이긴 게임수", win)
            
            game_tot = sum(tot_attempt)                            #전체 시도횟수의 합
            print("평균시도횟수", game_tot/(tot_game -1))
            
            sys.exit(0) #프로그램 종료
        inumber = int(number)    
        if ( n == inumber):
            print("정답!")
            win+=1
            
            tot_attempt.append(game)
            
            break
        elif ( n > inumber ):
            print("높은 수를 선택하시요")
        else:
            print("작은 수를 선택하시요")   
```

```python
#소득세 계산 : 소득 세율표를 구해보고 이를 이용하여 소득이 들어오면 세금이 얼마인지 계산해 보시오 ( 연봉. 단순히 세율만 적용합니다)

# 1천만원 미만시 연봉 9.5%/ 1천~4천: 19% / 4천~8천 : 28% / 8천이상 : 37%
```

```python
income = int(input("소득을 입력하시오:")) * 12
print("연봉:", income , end="만원 \n")

tax=0.0 #범위 지정해주면 round() 비슷?? # float형 초기화

if income<1000 :
    tax = income * 0.095
elif income<4000 :
    tax = 1000* 0.095 + (income - 1000)*0.019
elif income<8000 :
    tax = 1000*0.095 + 3000*0.19 + (income -4000)*0.28
else :
    tax = 1000*0.095 + 3000*0.19 + 4000*0.28 + (income - 8000)*0.37
    
print("총 세금은 " , round(tax,3), end="만원 입니다.")
    
```
