# for 중첩문에서
# 1차 - 내가 그리고 싶은 좌표 지정수 만큼의 크기

#i는 내가 지정해준 수 10로
#i는 0,10까지 정해졌고 ( 10줄이라는 공간 좌표로 생각했을때 ) 
# 2차 - print 하고자 하는 첫번째 결과값 : 여기는 공백
# 첫번째 for i in range( - i ) : range(0,10) 으로 들어가서
#k 에서 9까지 : 첫번째줄은 9 - i:0 ==> k 공백프린트를 9번해라
#k            : 두번째줄은 9 - i:1 ==> k 공백프린트를 8번해라

# 3차 - print 하고자 하는 두번째 결과값 : * 표기
# 두번째 for 문 중첩 + (i+1 )만큼 : i는 0 에서 9까지니까
#j는 0부터 9까지 : 첫번째줄은 0 + 1  ==> 별 * 1개 표기 

```python
#선생님 답


numberOfRows = int(input("숫자를 입력하시요: "))
for i in range(numberOfRows):
    for k in range(numberOfRows-i):
        print(" ", end="")
    for j in range(i+1):
        print("*", end="")
    for j in range(i):
        print("*", end='')
    print()
```
