```python
import random  
random.seed(1000) #seed() :
print(random.randint(1,100))


elements = ["earth", 1, "air", "fire", "water"]  
#다른언어에선 안되지만, 파이썬에서는 다른 자료형도 리스트화 가능, 왜냐면 주소니까
print(elements[0])


print(random.choice(elements)) # 내가 지정한 list 안에서 랜덤초이스 해라 ~
print(random.sample(elements, 2)) #샘플링 - R에서 임의추출
random.shuffle(elements)
print(elements)
```
