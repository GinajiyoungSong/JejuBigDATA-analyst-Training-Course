# 자료구조를 데이터 타입화

#### list : 데이터의 앞뒤에 다음이나 이전 데이터의 주소가 저장
         : 메모리를 효율적으로 사용하고 삽입삭제를 용이하게 하기 위해서 사용하는 자료구조
           검색에서는 속도가 느림                

#### dict : key값이 데이터가 저장되어지는 주소값으로 계산되는
####           hash 함수를 이용해 데이터 저장, 검색속도가 빠름
         데이터 주소값으로 변환되어지는 키이 : 데이터 형식으로 저장됨
         무순서.

#### set : key만 저장( 중복불허 ) : 범주형 데이터의 종류를 확인할 때 사용

#### tuple : list 와 동일한데 변경이 불가능함. (함수가 작음) 
####            치환하거나 추가 하는 함수가 아예 없음
       : 값을 전달하거나 여러개의 값을 리턴하고 싶을 때 사용 => 속도가 빠름
