# 사용자정의함수, class 함수 연습!

```python
class Student:
    studCount=0
    
    def __init__(self, name, kor, eng, mat):
        self.name=name
        self.kor=int(kor)
        self.eng=int(eng)
        self.mat=int(mat)
        self.tot=self.kor+self.eng+self.mat
        self.avg=self.tot/3
        
        Student.studCount+=1
        
# class Student 만들고        
                
```

```python
def start():
    students = []  # 빈 리스트를 만들것. 여러명의 데이터가 들어갈수 있게???
    
    while True :
        choice = input("1.입력 2.출력 3.계산 4.종료 =>")
        if choice =='1':
            name= input("이름:")
            kor=input("국어점수:")
            eng=input("영어점수:")
            mat=input("수학점수:")
            stud= Student(name, kor, eng ,mat) 
                              #클래스 정의한 Student를 instance 해서 객체. object 만듦
            students.append(stud) # 위에 초기화로 된 클래스변수들에 데이터를 넣어서 빈리스트에 넣기
        
        elif choice=='2': #출력기능
            for s in students:
                print("이름: %s 국어: %s 영어: %s 수학: %s 총점: %s 평균:%s" %(
                s.name, s.kor, s.eng, s.mat, s.tot, s.avg))
       
        elif choice=='3': # 계산하기 = 총점과 평균을 넣는 계산식을 넣어야한다
            kor_tot =0; eng_tot =0; mat_tot=0  
                      # 넣어줘야하는 데이터 초기화 시켜주고
                      
            for i in students: # 넣어준 학생의 데이터를 첫번째부터 끝까지 반복문에 넣는다
                kor_tot+=i.kor
                eng_tot+=i.eng
                mat_tot+=i.mat
            kor_avg = round(kor_tot /len(students), 2 ) # 소수점 2째자리까지 보려고 round 함수
            eng_avg = round(eng_tot /len(students), 2 )
            mat_avg = round(mat_tot /len(students), 2 )
            print("총 학생수 : %s 명" % len(students)) # students 에 1인분의 리스트 들어가니까
            print("국어평균 : %s\t\t 영어평균:%s\t\t 수학평균 :%s\t\t " % (
                kor_avg, eng_avg, mat_avg))
        
        elif choice=='4':
            break

start()
```

```python
import sys

# 1인분의 데이터를 넣는 성적표 클래스를 만들어라

class Student :
    
    def __init__(self) :
        self.bunho=0; self.name='';
        self.kor=0; self.mat=0; self.eng=0;
        self.total=0; self.average=0; self.grade=""  # input 데이터와 계산기로 넣어줄 데이터리스트
        
    def inputData(self):
        self.name=input("이름 :")
        self.kor=eval(input("국어점수 :"))
        self.mat=eval(input("수학점수 :"))
        self.eng=eval(input("영어점수 :"))
        
    def calcTotalAverage(self):
        self.total=self.kor + self.mat +self.eng
        self.average =round( self.total /3, 2)
        self.grade= self.calcGrade
        
    def calcGrade(self):
        if self.average >= 90 : return "A"
        elif self.average >=80 : return "B"
        elif self.average >=70 : return "C"
        elif self.average >=60 : return "D"
        else :                  return "F"
        
    def __str__(self):
        return "%1s   %5s   %5.2f   %5.2f   %5.2f   %5.2f   %5.2f   %1s" %(
        self.bunho, self.name, self.kor. self.mat, self.eng, self.total, self.average, self.grade)
    
    def __cmp__(self,other):
        return self.name == other.name
    
    
#1인분 데이터 넣는 성적 테이블을 만들었으니 이제 여러명을 넣을 데이터프레임 만들것
```

```python


class Management :
    schoolname = "제주아카데미"
    bunho= 0 #번호 초기화
    
    def __init__(self, count):
        print("메뉴를 선택하세요:")
        self.sungjuk=[]   # 클래스함수 하기 전에 일단 초기화해줘야함. 위에서 만든 성적표 넣을 리스트
        self.count=count   # 학생 1인분씩 추가될때마다 숫자 번호 오토로 넣기
        
    def input(self): # 번호 입력하는 카운트를 하기위해서, 초기화셋팅에 count 
        for count in range(self.count):
            stu = Student()  # 위에서 만든 Student 성적인스턴스 = 오브젝트해서 상속
            stu.inputData()  #위에 클래스함수 오브젝트.연결 => 데이터넣기
            Management.bunho +=1 # 위에 번호 초기화해둔거에 계속 추가해줘야함
            stu.bunho=Management.bunho # 카운트해서 학생번호에 넣는게 성적테이블 번호
            self.sungjuk.append(stu) #학생데이터들을 sungjuk 빈 리스트에 넣기
        
    def printSungjuk(self): # 출력하는 함수 3
        for per in self.sungjuk:
            print(per)
            print()
            
    def calcSungjuk(self): # 계산함수
        for per in self.sungjuk:
            per.calcTotalAverage() #위에서 정의한 나의 계산함수 상속....??
        
    def searchName(self): # 검색함수 => 내가 원하는 이름 호출
        name=input("검색할 학생 이름을 입력하세오:")
        for per in self.sungjuk:
            if per.name == name :
                print(per)
                print()
                return
        print('검색하는 학생이 없습니다')
        return
    
    def replaceName(self): # 이름수정함수 = 치환하기
        name=input("수정할 이름을 입력하세요: ")
        oldname = input('기존 이름 :') 
        
        for per in self.sungjuk:
            if per.name == oldname:
                per.name == name
                return
        print('수정할 이름이 없습니다')
        return
    
    def removeName(self): #이름삭제함수
        name=input("삭제할 이름을 적으세요:")
        
        for per in self.sungjuk:
            if per.name ==name:
                self.sungjuk.remove(per) # 검색한 이름에 해당되는 성적을 다 지움.
                return
        print("삭제되었습니다")
        return
    
    def sort(self): #정렬함수 
        return self.sungjuk.sort(key=lambda object : object.total, reverse = True)
    # 토탈 총점을 기준으로 가장 성적이 높은순) 내림차순 정렬하기.
        
```
