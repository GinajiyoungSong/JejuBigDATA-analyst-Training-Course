#정의함수의 연산을 가능하게 함

```python
# 파이썬 함수 - 리턴값의 지정은 없음
# define 함수를 정의한다
def calc(a,b):         #함수의 역할 : 작업, 매개변수 : 작업을 할 때 필요한 데이터
    return a+b, a*b   # return 값 2개이상 가능한게 파이썬!
                      # 내가 지정한 사용자 정의함수를 이용해서
x,y = calc(3,4)       # x= 첫번째 리턴값 a+b,   y= 두번째 리턴값 a*b
print(x,y)
print(x)


x,y
args=(4,5)   # () : tuple 데이터가 되었음.
calc(*args) #포인터 < 지정함수( *데이터) > 
```

```python
number=int(input("수를 입력하시오"))
           
def myfunc(num):
    if num%3==0:
        print("가위")
    elif num%3==1:
        print("바위")
    else:
        print("보")
           
           
myfunc(number)
```

## 문자열 거짓 : 아무것도 없을때 ' ' 공백도 문자열
## 숫자의 거짓 : 0
## list의 거짓:[]
## dict의 거짓 : {}
## tuple의 거짓 : ()

```python
# 날짜의 입력은 문자열
# 특수하게 해석해서 년월일로 해석

import datetime
dt = datetime.datetime.now()   #함수는 () 사용해야함
print(dt)
print()

# 데이터입력
age=input("생년월인(19951031)을 입력하시오")

# 데이터 처리
year = int(age[0:4]) # 인덱싱 위치로 read    #마지막은 미포함
month = int(age[4:6])
day = int(age[6:])

# 빼기 연산
# 분산 : 상계방지 (제곱)
year_diff = dt.year - year
month_diff = dt.month - month
day_diff = dt.day -day
if day_diff < 0 :  # 일 내림 처리
    month_diff -=1
if month_diff <0 : # 월 내림 처리
    year_diff -=1
print("당신의 나이는", year_diff + 1)
if year_diff < 18 :
    print("당신은 미성년자입니다.")
else :
    print("당신은 미성년자가 아닙니다.")
```

## 문제 1 ) 제곱미터로 입력되는 아파트 면적을 평형으로 계산하여 

## 종류를 판정하시오

#(30평을 기준으로 판정)하는 프로그램 작성
#소형(17) 중소형(25) 중형 (30)
#평형수 = 제곱미터 / 3.305


```python
area = int(input("면적을 입력하시오"))
room = area/3.305

print("아파트면적:",round(room,3), end="평형")

# 입력받은 수를 평형으로 계산 하였으니, 이제 종류를 판정하는 조건문을 만들것
if room>=30:
    print('아파트는 30평형 이상으로 대형입니다')
elif 30> room>=25:
    print('아파트는 25평형 이상이나 30평형 미만으로 중형입니다')
elif 25> room>=17:
    print('아파트는 17평형 이상이나 25평형 미만으로 중소형입니다')
else : print('아파트는 17평형 미만으로 소형입니다')
```

#1천만원 미만시 연봉 9.5%/ 1천~4천: 19% / 4천~8천 : 28% / 8천이상 : 37%


## 놀이공원 매표소 단체 입장료 계산하기

### 조건 : 팀별 인원 구성을 입력

#( 초등, 청소년, 일반인, 경로대상) 
#정산요금 : 초등(5000원), 청소년(10000원), 일반인(15000원), 경로대상(3000원)

#팀별 할인 카드 확인
#카드없음 : 할인 무, 일반카드 : 10% , vip : 20%

```python
ele=int(input("초등학생은 몇명입니까"))
under19=int(input("청소년은 몇명입니까"))
adult=int(input("일반인은 몇명입니까"))
old=int(input("경로대상 어른은 몇명입니까"))
card = int(input("할인카드가 있습니까? 없으면 :0, 일반카드:1, vip카드:2 를 입력하시오 "))
calc= ele*5000 + under19*10000 + adult*15000 + old*3000

#만약 할인카드가 있다면

if card== 2:
    calc*0.8
elif card ==1:
    calc*0.9
else card ==0:
    calc*1
    
print("총 계산요금은 :", round(calc,2), end="원입니다")
    
```
