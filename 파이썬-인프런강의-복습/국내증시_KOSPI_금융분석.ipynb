```
import FinanceDataReader as fdr


# KOSPI index, 2019 코로나 발~ Now
ks11 = fdr.DataReader('KS11', '2019-12-01')

# Dow Jones Industrial(DJI), 2019.12 ~ Now
dji = fdr.DataReader('DJI', '2019-12-01')

# USD/KRW, 2019-12-01 ~Now
usdkrw = fdr.DataReader('USD/KRW', '2019-12-01')

# KRX stock symbol list and names
krx = fdr.StockListing('KRX')

# KRX stock delisting symbol list and names 상장폐지 종목 전체 리스트
krx_delisting = fdr.StockListing('KRX-DELISTING')
```

```
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (14,8)
plt.plot(ks11['Open'])
plt.plot(ks11['Close'])
plt.legend(['Opne', 'Close'])
```

```
plt.rcParams["figure.figsize"] = (14,8)
plt.plot(ks11['High'], color='green')
plt.plot(ks11['Low'], color='red')
plt.legend(['High', 'Low'])
```
