```python
%matplotlib inline 
# 쥬피터에서만 사용
import matplotlib.pyplot as plt # 시각화

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

import mglearn # 학습도움

# MLP  Multi Layer Perceptron : FFNN(Feed forward neural network)
# 멀티레이어 쓰는 이유 => XOR (기울기소멸)문제를 해결, 레이어가 많으면 정밀도 자동으로 높아진다.
from sklearn.neural_network import MLPClassifier #MLP Regressor
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
X, y= make_moons(n_samples=100, noise=0.25, random_state=3)
```

```python
                                                       # stratify 층화 변수 선택법
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42)
mlp = MLPClassifier(random_state=0).fit(X_train, y_train)
# mpl 모델에 fit 시켰음
# 위 모델에 적합하게 가중치 완성 ->
# 레이어 지정안해서 => Multi Layer Perceptron 가중치 레이어 사이즈자동으로 해줌

mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("feature 0")
plt.ylabel("feature 1")
```

```python
 'activation': 'relu' : 0 이하를 제거한 활성화 함수 (activation function) : 속도가 빨라짐
 'learning_rate': 'constant' 변하지않음
 'learning_rate_init': 0.001 (초기화값)
  'max_iter': 200, 반복횟수 = epoch 1회반복을 말함
 'solver': 'adam' (그래프 보면 일반화되는게 보임)
    learing-rate 를 조절, momentum을  사용하는  = optimizer 
( 지역해를 넘기 위해서 가는 방향으로 더 넘겨보기 <- 처음에는 많이주고 점점 줄여주기 (진동을) ) 
```

```python
Black Box Test는
1. 내부 구조가 어떻게 되어있는지 확인할 수 없다.
2. 결과물 상태를 보고 요구사항이 충족하는지 확인한다.
3. 테스트 비용이 많이 든다. 왜냐면 어떠한 결과가 나올지 예측할 수 없기 때문에 test case를 많이 시도해보아야 하기 때문이다.

 

White Box Test는
1. 소스코드 형태로 소프트웨어를 테스트한다.
2. 테스트 코드는 White Box Test에 해당하는 것 같다. Unit Testing, Integration Testing, System Testing이 해당된다.
3. 시작단계부터 테스트를 진행할 수 있다. 테스트 비용도 적다.
```

```python
white는 프레임워크 혹은 컴포넌트의 내부를 깊게 들여다볼 수 있도록 되어 있어서, 
객체의 구현 내부에 접근하는 코드를 쓸 수 있다. 자유도가 더 높겠지만, 기존 의도대로 동작하지 않을 수도 있고, 
다른 프레임워크나 프로그램과 맞지 않을 수도 있다는 단점이 있겠다.

 

black은 컴포넌트 안을 보거나 변경할 수 없으므로 고안된 대로 재사용이 가능하다. 
내부 알고리즘을 알 수 없지만, 고안된 대로만 쓸 수 있기 때문에 일관된 프로그래밍이 가능하겠다.

# Black Box => 원인규명

```

```python
fig, axes = plt.subplots(2, 4, figsize=(20,8))
for  axx, n_hidden_nodes in zip(axes, [10,100]): # 히든레이어 은닉층을 첫번째층에서 10개 3번째층에서 100개
    # 결과값으로 10차 은닉층 보다는 100개의 퍼셉트론 은닉층이 더 많이 활성화함수를 거쳐서 가중치합이 fix 됨! = 10차원과 100차원차이
    
    for ax, alpha in zip(axx, [0.1, 0.001, 0.0001, 0.00001]): # alpha는 규제값 => 규제가 작아질수록 과적합되고 있음
        mlp = MLPClassifier(solver='lbfgs', random_state=0,
                          activation='tanh',
                          hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes], alpha=alpha)
        
        # 반복문에 알파값=규제값을 조정해서 다양한 조건의 모델을 만듦
        mlp.fit(X_train, y_train)
        # mlp 는 비선형회귀 => 다항회귀 (다중회귀는 변수가 많다는 뜻)
        
        
        
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
        ax.set_title("n_hidden=[{}, {}]\nalpha={:.4f}".format(n_hidden_nodes, n_hidden_nodes, alpha))
```

```python
S = mglearn.datasets.make_signals()
plt.figure(figsize=(6,1))
plt.plot(S, '-')
plt.xlabel("시간")
plt.ylabel("신호")
plt.margins(0)
```

```python
import numpy as np
# 노이즈 : 전송 데이터 = 전송중에 노이즈가 섞임
A = np.random.RandomState(0).uniform(size=(100,3)) # 랜덤으로 뽑아서 합집합으로 강제 노이즈 처리
X = np.dot(S, A.T)
print("측정 데이터 형태: {}".format(X.shape))
plt.figure(figsize=(14,2))
plt.plot(A, '-')
```

```python
from sklearn.decomposition import NMF, PCA
nmf = NMF(n_components=3, random_state=42)
S_ = nmf.fit_transform(X)
print("복원한 신호 데이터 형태: {}".format(S_.shape))
```

```python
pca = PCA(n_components=3)
H = pca.fit_transform(X)
```

```python

#       원본, 노이즈, NMF, PCA
models = [S, X, S_, H] # PCA 는 노이즈: 양수 데이터 노이즈를 제거못함
names = ['측정 신호 ( 처음 3개)', '원본 신호', 'NMF로 복원한 신호', 'PCA로 복원한 신호']

fig,axes = plt.subplots(4, figsize = (20, 10), gridspec_kw={'hspace':.5},
                       subplot_kw={'xticks':(), 'yticks':()})

for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:, :3], '-')
    ax.margins(0)
```
