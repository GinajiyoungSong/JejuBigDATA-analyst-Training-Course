
## vi편집기 - 리눅스에서 가장많이 사용하는 덱스트 에디터

```python
입력모드는 메모장처럼 텍스트를 자유롭게 편집하는 모드
명령모드는 말그대로 다양한 명령을 내리는 모드
처음 vi편집기를 실행하면 명령모드로 시작

-명령모드 : 입력모드에서 명령모드로 전환
           -esc키를 누른다 + 커서가 있는 곳에서 dd(한줄지움), yy(복사), p(붙여넣기)

명령모드에서 입력모드로 전환
 -입력모드  - 커서앞에서 i (insert), a(append) , o(빈행추가하여 입력) , s(글자를 지우고 입력)
    
    
-마지막행 모드 = :w (저장) , :q (종료) :q! (저장하지 않고 종료) :wq (저장하고 종료)
```
