# 객체 지향 프로그래밍 ( class(끝) => (instance: 메모리확보), object(객체) )

-class : 멤버변수(속성) + 멤버함수
    - 인스턴스할때는 멤버변수(속성)를 저장하는 공간만 확보
    
    - 멤버함수는 계속 만들어지는게 아니라 공유되어짐.

```python
width = 10
height = 20
area = 10 * 20/2
print("삼각형의 넓이는", area) #이렇게 출력해줘야만 뭔지 알수 있음
```

```python
class Triangle:
    def setData(self, width, height): # 일반 함수와 달리 함수를 이용한 초기화를 해줘야함
        self.width =width
        self.height=height
    def area(self):                           #함수를 만들때 (self) 넣어줘야함 #매개변수 취급은 안함
        return self.width*self.height/2
    
    
# 인스턴스 한다. : triangle 초기화    tri1 이라는 변수로 ()함수처럼 만듦
tri1 = Triangle()                        
tri1.setData(10,20)   #실시간 변수확보.멤버변수 저장하는 변수공간 확보
print(tri1.width, tri1.height, tri1.area())
```

```python
#__init__: object 에 있는 함수 ( 상속 : 클래스를 생성하면 자동으로 상속)

# 위에 만든 object 를 상속한것
# tri = Triangle() 인스턴트 했을때 이미 object 만드니까

class Triangle :
    def __init__(self, width, height):           #초기화함수 ,1 생성자. 강제로 호출불가한 함수
        self.width =width
        self.height=height
    def area(self):                           #함수를 만들때 (self) 넣어줘야함 #매개변수 취급은 안함
        return self.width*self.height/2
    
tri1 = Triangle(10,20)                        
# tri1.__init__(10,20) 불가함 = 인스턴트할때 자동으로 실시간 매개변수 저장 : 변수공간도 확보함

print(tri1.width, tri1.height, tri1.area())
```

```python
# 클래스변수, 멤버변수, 지역변수
# 상속받을때 class 클래스함수이름(object) <-파이썬에서 이미 만들어진 object 함수
#                                -------- 이부분 생략가능

class MyClass(object):              #선언하는 순간에 메모리에 자리를 잡음
    """클래스의 예"""
    i = 12345                   # 함수 바깥에서 선언되어 지는 변수는 클래스변수 , 공용변수
    def __init__(self):
        self.i=54321            # self. 가 붇은 변수가 멤버변수
        num=100                 # self 없는게 지역변수
    def f(self):
        return self.i
MyClass.__doc__                 #object의 속성


# 함수가 실행을 종료하더라도 멤버변수는 없어지지 않음   self.i=54321   이건 계속 남음
```

```python
from time import time, ctime, sleep #시간은 초로 관리  :ctime ( 문자열로convert)
class Life:
    def __init__(self):           #생성자 
        self.birth=ctime() # 현재의 시간을 문자열로 변환
        print('생성', self.birth)
    def __del__(self):            #소멸자 (메모리에서 사라질 때 자동으로 호출 )
        print('사망', ctime())     # 시스템에서 참조함: 프린트해주고 바로 삭제~~~~
    def test(self):
        #mylife =Life()
        print('Sleepint for 3 sec')
        sleep(3) # 초
    
li = Life() # 오른쪽은 heap 에 저장, 왼쪽은 그 주소를 가리키는 stack 에 생성
li.test()
del li     #  del : 참조를 지워버리면 자동으로 garbege 컬렉션 됨
```

```python
# 사원관리
class Employee:
    empCount = 0 #사원수를 저장 : 클래스 변수 생성
    def __init__(self, name, salary):              #초기화 함수가 있기때문에 아래서  이름, 급여 넣음
        self.name = name
        self.salary = salary # 급여
        Employee.empCount +=1
    def displayCount(self):
        print( "직원 수= %d" % Employee.empCount)
    def displayEmployee(self):
        print(" 이름 : ",self.name , ", 급여 : ", self.salary)
    def __call__(self, *pargs , **kargs):
        print(" Called:", pargs, kargs) # 클래스가 함수처럼 작동?????
        
```

```python
import math
#1개를 다루는 클래스를 작성하고 있지만 거리값은 두개를 사용!

class Point:
    def __init__(self, x, y): # 생성자
        self.x=x    #매개변수 x,y를 넣어주면서 맴버변수 초기화 self.x, self.y 이녀석들
        self.y=y
    def __str__(self): # __repr__ 문자열을 요구하는 함수에 들어가면
        return "좌표는 (%d, %d)" %(self.x, self.y)
    
    def dist(self, other): #피타고라스 정리
        distance=math.sqrt((self.x-other.x)**2 + (self.y-other.y)**2)
        return distance
    def collision(self, other):
        return self.dist(other) <3
```

```python
first =Point(7,3)
second =Point(4,7)
print(first)
print(second)
```

```python
print(" 두점 사이의 거리는 =", first.dist(second))
if first.collision(second):
    print("충돌했습니다")
else :
    print("충돌하지 않았습니다")
```

# 문제 : Point 클래스를 이용하여 Circle 클래스를 정의하시오
- 원 : 중심점(point)하고 반지름
- 원과 원이 충돌하는 여부를 확인하는 함수를 작성하시요

```python
# center =Point() 함수에 x 값으로 들어가고 매개변수가 radius 추가됨
class circle:
    def __init__(self, center, radius): 
        self.center=center   
        self.radius=radius
    def __repr__(self): 
        return "원의 중심은" + str(self.center) + "반지름은" + str(self.radius)
    def collision(self,other):
        if self.center.dist(other.center) < self.radius + other.radius:
            result = "충돌한다"
        else:
            return "충돌하지 않는다"
        return result
        
```

```python
cir1=circle(first,1)
cir2=circle(second,5)

print(cir1)
print(cir2)

print(cir1.collision(cir2))
circle.collision(cir1,cir2) #self
```

```python
# 문제 : vector class를 작성하시요
- x,y 의 2차원 벡터 구현
- 벡터의 연산 (+,-,*,/): 요소별로 계산하면 됨
- dot 함수를 구현하시요 : 벡터의 요소별로 곱하고 모두 더해주면 내적 유사도
- cos 값으로 변환하는 함수를 구현하시오


    cos theta = acos(A 내저규 / |A||B|)
           |A| = math.sqrt(x^2 + y^2)
```

```python
class Vector:
    def __init__(self, x, y):
        self.x =x
        self.y =y
    def __repr__(self):
        return "좌표는 (%d, %d)" % (self.x, self.y)
    def __add__(self,other): #연산자 오버로딩을 오버라이딩으로 구현한 것
        return Vector(self.x + other.x, self.y + other.y)
    def __mul__(self,other):
        return  Vector(self.x * other.x, self.y * other.y)
    def __sun__(self,other):
        return Vector(self.x - other.x, self.y - other.y)
    def __truediv__(self,other): # div : 2버전에서 사용
        return Vector(self.x / other.x, self.y / other.y)
```

```python
f = Vector(10,23)
g = Vector(20,27)
print(f)
print(f+g)
print(f*g)
print(f/g)
```

```python
# 3 차원 벡터를 클래스로 구현하시요 ( 연산자 오버로딩 +, - , *, /)
```

```python
class Vector:
    def __init__(self, x, y, z):
        self.x =x
        self.y =y
        self.z =z
    def __repr__(self):
        return "좌표는 ( %d, %d, %d)" % (self.x, self.y, self.z)
    def __add__(self,other): #연산자 오버로딩을 오버라이딩으로 구현한 것
        return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
    def __mul__(self,other):
        return  Vector(self.x * other.x, self.y * other.y, self.z - other.z)
    def __sun__(self,other):
        return Vector(self.x - other.x, self.y - other.y,  self.z * other.z)
    def __truediv__(self,other): # div : 2버전에서 사용
        return Vector(self.x / other.x, self.y / other.y,  self.z / other.z)
    def dist(self):
        return math.sqrt(self.x*self.x+self.y*self.y+self.z*self.z)
    # 방향만 바뀌어야지 크기까지 바뀌면 안됨
    
    
    
    
    def norm(self): #정규화(normarization): 크기값이 1인 사이즈로 변경( 반지름이 1인 )
        dist=self.dist()
        return self.x/dist, self.y/dist, self.z/dist
    
    def dot(self, other): #내적, (두벡터의 사이각과 같은 것)
        return self.x*other.x + self.y*other.y + self.z* other.z
    
    
    def theta(self, other):
        dot = self.dot(other) # 행렬의 곱셈. 여기서는 3차원이상의 곱의 합들???
        dist1 =self.dist()
        dist2 =other.dist() #radian => degree 로, (컴퓨터는 라디안 만 사용)
        return math.acos(dot/(dist1*dist2)) * (180/math.pi)#라디안 => degree로 변경 :pi 로 나눔
    
    
    
    # 변환 과정이 필요 => 사람이 사용하는 각도 입력(90) => 라디안 변환(math.pi/180)
    
    # 계산 후 => 사람이 사용하는 degree 로 변환하여 출력 =>180/math.pi
```

```python
#1 라디안, 각의 크기를 재는 SI 단위.
# 기호는 rad 원의 둘레 위에서의 반지름의 길이와 같은 길이의 호를 절취한 2개의 반지름 사이의 포함되는 평면각
# 약 57.3도

# 우리가 일상적으로 사용하는 각의 단위 degree

#컴퓨터나 수학에서 주로 사용되는 각 체계는 radian
# 사용자에게 각을 입력받을 때는 보통 degree 로 입력받은 후 이를 내부적으로 radian 으로 고쳐서 계산
# 180 degree =  파이 radian
# 1 degree = 파이 / 180 radian
```

```python
f= Vector(10,0,0)
g= Vector(0,27,0)

print(f)
print(f + f)
print(f + g)
print(f * g)
print("f벡터의 크기 =", f.dist())
print("g벡터의 크기=", g.dist())
print("f 의 normal vector", f.norm()) #1.0 0.0 0.0
print("g 의 normal vector", g.norm()) #0.0 1.0 0.0
print(" 두벡터의 내적 =", g.dot(f)) #0   내적이 0이명 두 벡터는 90도 직교한다
print("f와 g의 사이각은 =", f.theta(g)) #90
```

```python
# 이름, 국어, 수학, 영어를 관리하는 student class 를 작성하시오 (1인분 관리)
- 총점과 평균도 계산해서 멤버 변수에 지정
```

```python
class Student:
    studCount=0
    def __init__(self, name, kor, eng, mat):
        self.name=name
        self.kor=int(kor)
        self.eng=int(eng)
        self.mat=int(mat)
        self.tot=self.kor+self.eng+self.mat
        self.avg =self.tot / 3
        Student.studCount+=1
        
```

```python
# 메뉴화
def start():
    students =[]
    while True :
        choice = input("1.입력 2.출력 3.계산 4.종료 =>")
        if choice =='1':
            name=input("이름 :")
            kor=input("국어점수 :")
            eng=input("영어점수 :")
            mat=input("수학점수 :")
            stud =Student(name, kor, eng, mat)
            students.append(stud)
        elif choice=='2':
            for s in students:
                print("이름: %s 국어: %s 영어: %s 수학: %s 총점: %s 평균: %s" %(
                s.name, s.kor, s.eng, s.mat, s.tot, s.avg))
        elif choice== '3':
            kor_tot =0; eng_tot =0; mat_tot=0
            for i in students:
                kor_tot+=i.kor
                eng_tot+=i.eng
                mat_tot+=i.mat
            kor_avg = round(kor_tot /len(students),2 )
            eng_avg = round(eng_tot /len(students),2 )
            mat_avg = round(mat_tot /len(students),2 )
            print("총 학생수 : %s 명" % len(students))
            print("국어평균 : %s\t\t 영어평균: %s\t\t 수학평균 :%s\t\t " %(
                 kor_avg, eng_avg, mat_avg))
        elif choice== '4':
            break
            
            
start()
```

```python
import sys

# 1인분의 데이터를 처리하는 class 를 만든다

class Student :
    def __init__(self) :
        self.bunho=0; self.name='';
        self.kor=0; self.mat=0; self.eng=0;
        self.total=0; self.average=0; self.grade=""
    def inputData(self):     
        self.name=input("이름 :")
        self.kor=eval(input("국어점수 :"))
        self.mat=eval(input("수학점수 :"))
        self.eng=eval(input("영어점수 :"))
    def calc_total_average(self):
        self.total =self.kor + self.mat + self.eng
        self.average =round( self.total / 3, 2)
        self.grade= self.calcSemGrade()
    def calcSemGrade(self):
        if self.average >=90: return "A"
        elif self.average >=80: return "B"
        elif self.average >=70: return "C"
        elif self.average >=60: return "D"
        else:                   return"R"
    def __str__(self):
        return "%5s  %5s  %6.2f  %6.2f  %6.2f  %7.2f  %7.2f  %5s" % (
            self.bunho, self.name, self.kor, self.mat, self.eng, self.total, self.average, self.grade)
    def __cmp__(self,other): # compare, true/false 로 결과 리턴
        return self.name == other.name
    
```
