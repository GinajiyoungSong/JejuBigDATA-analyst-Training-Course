```python
class Triangle:
    # 일반 함수와 달리 함수를 이용한 초기화를 해줘야함
    # __init__ 내장 함수 안쓰고  = 사용자 정의함수를 만들어줬을때
    def setData(self, width, height): 
        self.width =width
        self.height=height
       
    #함수를 만들때 (self) 넣어줘야함 #매개변수 취급은 안함
    def area(self):                           
        return self.width*self.height/2
    
    
# 인스턴스 한다. : triangle 초기화    tri1 이라는 변수로 ()함수처럼 만듦

tri1 = Triangle()                        
tri1.setData(10,20)   #실시간 변수확보.멤버변수 저장하는 변수공간 확보
print(tri1.width, tri1.height, tri1.area())
```

```python
import math
#1개를 다루는 클래스를 작성하고 있지만 거리값은 두개를 사용!

class Point:
    
    def __init__(self, x, y): # 생성자크
       #매개변수 x,y를 넣어주면서 맴버변수 초기화 self.x, self.y 이녀석들
        self.x=x    
        self.y=y
        
        
    def __str__(self): # __repr__ 문자열을 요구하는 함수에 들어가면
        return "좌표는 (%d, %d)" %(self.x, self.y)
    
    def dist(self, other): #피타고라스 정리
        distance=math.sqrt((self.x-other.x)**2 + (self.y-other.y)**2)
        return distance
    
    
    def collision(self, other):
        return self.dist(other) <3
```

```python
first =Point(7,3)
second =Point(4,7)
print(first)
print(second)
```

```python
print(" 두점 사이의 거리는 =", first.dist(second))
if first.collision(second):
    print("충돌했습니다")
else :
    print("충돌하지 않았습니다")
```

# 문제 : Point 클래스를 이용하여 Circle 클래스를 정의하시오
- 원 : 중심점(point)하고 반지름
- 원과 원이 충돌하는 여부를 확인하는 함수를 작성하시요

```python
# center =Point() 함수에 x 값으로 들어가고 매개변수가 radius 추가됨
class circle:
    def __init__(self, center, radius): 
        self.center=center   
        self.radius=radius
    def __repr__(self): 
        return "원의 중심은" + str(self.center) + "반지름은" + str(self.radius)
    def collision(self,other):
        if self.center.dist(other.center) < self.radius + other.radius:
            result = "충돌한다"
        else:
            return "충돌하지 않는다"
        return result
        
```

```python
cir1=circle(first,1)
cir2=circle(second,5)

print(cir1)
print(cir2)

print(cir1.collision(cir2))
circle.collision(cir1,cir2) #self
```

# 문제 : vector class를 작성하시요


- x,y 의 2차원 벡터 구현
- 벡터의 연산 (+,-,*,/): 요소별로 계산하면 됨

- dot 함수를 구현하시요 : 벡터의 요소별로 곱하고 모두 더해주면 내적 유사도
- cos 값으로 변환하는 함수를 구현하시오


    cos theta = acos(A 내적 / |A||B|)
           |A| = math.sqrt(x^2 + y^2)

```python
class Vector:
    def __init__(self, x, y):
        self.x =x
        self.y =y
        
    def __repr__(self):
        return "좌표는 (%d, %d)" % (self.x, self.y)
    
    def __add__(self,other): #연산자 오버로딩을 오버라이딩으로 구현한 것
        return Vector(self.x + other.x, self.y + other.y)
    
    def __mul__(self,other):
        return  Vector(self.x * other.x, self.y * other.y)
    
    def __sub__(self,other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __truediv__(self,other): # div : 2버전에서 사용
        return Vector(self.x / other.x, self.y / other.y)
```

```python
f = Vector(10,23)
g = Vector(20,27)
print(f)
print(f+g)
print(f*g)
print(f/g)
```

# 3 차원 벡터를 클래스로 구현하시요 ( 연산자 오버로딩 +, - , *, /)

```python
class Vector:
    def __init__(self, x, y, z):
        self.x =x
        self.y =y
        self.z =z
        
    def __repr__(self):
        return "좌표는 ( %d, %d, %d)" % (self.x, self.y, self.z)
    def __add__(self,other): #연산자 오버로딩을 오버라이딩으로 구현한 것
        return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
    
    def __mul__(self,other):
        return  Vector(self.x * other.x, self.y * other.y, self.z - other.z)
    
    def __sub__(self,other):
        return Vector(self.x - other.x, self.y - other.y,  self.z * other.z)
    
    def __truediv__(self,other): # div : 2버전에서 사용
        return Vector(self.x / other.x, self.y / other.y,  self.z / other.z)
    
    def dist(self):
        return math.sqrt(self.x*self.x+self.y*self.y+self.z*self.z)
    # 방향만 바뀌어야지 크기까지 바뀌면 안됨
    
    
    
    
    def norm(self): #정규화(normarization): 크기값이 1인 사이즈로 변경( 반지름이 1인 )
        dist=self.dist()
        return self.x/dist, self.y/dist, self.z/dist
    
    def dot(self, other): #내적, (두벡터의 사이각과 같은 것)
        return self.x*other.x + self.y*other.y + self.z* other.z
    
    
    def theta(self, other):
        dot = self.dot(other) # 행렬의 곱셈. 여기서는 3차원이상의 곱의 합들???
        dist1 =self.dist()
        dist2 =other.dist() #radian => degree 로, (컴퓨터는 라디안 만 사용)
        return math.acos(dot/(dist1*dist2)) * (180/math.pi)#라디안 => degree로 변경 :pi 로 나눔
    
    
    
    # 변환 과정이 필요 => 사람이 사용하는 각도 입력(90) => 라디안 변환(math.pi/180)
    
    # 계산 후 => 사람이 사용하는 degree 로 변환하여 출력 =>180/math.pi
```

```python
#1 라디안, 각의 크기를 재는 SI 단위.
# 기호는 rad 원의 둘레 위에서의 반지름의 길이와 같은 길이의 호를 절취한 2개의 반지름 사이의 포함되는 평면각
# 약 57.3도

# 우리가 일상적으로 사용하는 각의 단위 degree

#컴퓨터나 수학에서 주로 사용되는 각 체계는 radian
# 사용자에게 각을 입력받을 때는 보통 degree 로 입력받은 후 이를 내부적으로 radian 으로 고쳐서 계산
# 180 degree =  파이 radian
# 1 degree = 파이 / 180 radian
```

```python
# 이름, 국어, 수학, 영어를 관리하는 student class 를 작성하시오 (1인분 관리)
- 총점과 평균도 계산해서 멤버 변수에 지정
```

```python
class Student:
    studCount=0 #공용변수
    
    def __init__(self, name, kor, eng, mat):
        self.name=name                            # 멤버변수
        self.kor=int(kor)
        self.eng=int(eng)
        self.mat=int(mat)
        self.tot=self.kor+self.eng+self.mat
        self.avg =self.tot / 3
        Student.studCount+=1                    # self 없으면 지역변수 
        
# class Student 만들고      
```

```python
# 메뉴화 = > 사용자 정의함수로

def start():
    students = []  # 빈 리스트를 만들것. 여러명의 데이터가 들어갈수 있게???
    
    while True :
        choice = input("1.입력 2.출력 3.계산 4.종료 =>")
        if choice =='1':
            name= input("이름:")
            kor=input("국어점수:")
            eng=input("영어점수:")
            mat=input("수학점수:")
            stud= Student(name, kor, eng ,mat) 
                              #클래스 정의한 Student를 instance 해서 객체. object 만듦
            students.append(stud) # 위에 초기화로 된 클래스변수들에 데이터를 넣어서 빈리스트에 넣기
        
        elif choice=='2': #출력기능
            for s in students:
                print("이름: %s 국어: %s 영어: %s 수학: %s 총점: %s 평균:%s" %(
                s.name, s.kor, s.eng, s.mat, s.tot, s.avg))
       
        elif choice=='3': # 계산하기 = 총점과 평균을 넣는 계산식을 넣어야한다
            kor_tot =0; eng_tot =0; mat_tot=0  
                      # 넣어줘야하는 데이터 초기화 시켜주고
                      
            for i in students: # 넣어준 학생의 데이터를 첫번째부터 끝까지 반복문에 넣는다
                kor_tot+=i.kor
                eng_tot+=i.eng
                mat_tot+=i.mat
            kor_avg = round(kor_tot /len(students), 2 ) # 소수점 2째자리까지 보려고 round 함수
            eng_avg = round(eng_tot /len(students), 2 )
            mat_avg = round(mat_tot /len(students), 2 )
            print("총 학생수 : %s 명" % len(students)) # students 에 1인분의 리스트 들어가니까
            print("국어평균 : %s\t\t 영어평균:%s\t\t 수학평균 :%s\t\t " % (
                kor_avg, eng_avg, mat_avg))
        
        elif choice=='4':
            break

start()
```

```python
import sys

# 1인분의 데이터를 처리하는 class 를 만든다

class Student :
    def __init__(self) :
        self.bunho=0; self.name='';
        self.kor=0; self.mat=0; self.eng=0;
        self.total=0; self.average=0; self.grade=""
        
    def inputData(self):     
        self.name=input("이름 :")
        self.kor=eval(input("국어점수 :"))
        self.mat=eval(input("수학점수 :"))
        self.eng=eval(input("영어점수 :"))
        
    def calcTotalAverage(self):
        self.total =self.kor + self.mat + self.eng
        self.average =round( self.total / 3, 2)
        self.grade= self.calcGrade()
        
    def calcGrade(self):
        if self.average >=90: return "A"
        elif self.average >=80: return "B"
        elif self.average >=70: return "C"
        elif self.average >=60: return "D"
        else:                   return"F"
        
    def __str__(self):
        return "%5s   %5s    %6.2f    %6.2f    %6.2f    %7.2f    %7.2f    %5s" % (
            self.bunho, self.name, self.kor, self.mat, self.eng, 
            self.total, self.average, self.grade)
    
    def __cmp__(self,other): # compare, true/false 로 결과 리턴
        return self.name == other.name
    
```

```python
# 여러명을 관리하기 위해서,
# 위에 1인분의 관리 class 함수를 가져다가 씀

class Management :
    schoolname = "제주 아카데미"     #공용변수
    bunho = 0
    
    
    def __init__(self, count):      #생성자
        print("메뉴를 선택하시오")
        self.sungjuk=[]
        self.count=count
        
        
    def input(self):                           # 1) 입력
        for co_in in range(self.count):
            stu = Student()                    # Student 클래스를 로딩 = 위의 함수를 다 이용
            
            stu.inputData()                    # 위의 멤버함수 상속
                                               # stu.inputData() 함수 이용 데이터넣고
                
            Management.bunho +=1              # 번호를 자동으로 입력하기 위해
            stu.bunho = Management.bunho
            self.sungjuk.append(stu)
            
            
    def print_sungjuk(self):
        for per in self.sungjuk:                #2) 출력 :sungjuk [] 만든 리스트에서 하나씩 확인
            print(per)
            print()
            
   
                                                #3)계산 sungjuk [] 만든 리스트에서 하나씩 확인
    def calc_sungjuk(self): # 계산함수
        for per in self.sungjuk:
            per.calcTotalAverage() 
            
            
            
    def search_name(self):
        name=input("검색할 학생 이름을 입력하시오")    # 검색  => 내가 원하는 이름 호출
        for per in self.sungjuk:                      #sungjuk [] 만든 리스트에서 하나씩 확인
            if per.name == name :
                print(per)
                return
        print("검색하고자 하는 학생이 없습니다.")
        return
    
    
    def rename(self): # 이름수정함수 = 치환하기
        name=input("수정하고자 하는 학생 이름을 입력하시오")
        for per in self.sungjuk:
            if per.name == name :
                newname =input("수정할 이름을 추가하시오")
                per.name = newname
            return
    
    def delete(self):
        name = input(" 삭제하고자 하는 이름을 입력하시오")
        for per in self.sungjuk:
            if per.name ==name:
                self.sungjuk.remove(per) # 검색한 이름에 해당되는 성적을 다 지움.
                return
        print("삭제되었습니다")
        return
    
    def sort(self): #정렬함수 
        return self.sungjuk.sort(key=lambda object : object.total, reverse = True)
    # 토탈 총점을 기준으로 가장 성적이 높은순) 내림차순 정렬하기.
        
        
        
        
```
