```python
import pymysql
# 네트워크 접속
conn = pymysql.connect(host='192.168.1.248',
                      user='root', port=3306, password='123', db='acornschool',
                      charset='utf8mb4', use_unicode=True, cursorclass=pymysql.cursors.DictCursor)

cur = conn.cursor()

# 만든 테이블에 데이터 value 넣기
# 1. 대표문을 아예 만들어서 데이터를 넣는 방법
sql = '''INSERT INTO pet (name, owner, species, sex, birth, death)
      VALUES(%s, %s, %s, %s, %s, %s)'''
cur.execute(sql, ('유리', 'acorn', '고양이', 'f', '2014-1-1', '2016-2-1'))
cur.execute(sql, ('귀염', 'acorn', '강아지', 'f', '2014-3-1', '2016-9-1'))
cur.execute(sql, ('돌이', 'acorn', '곰', 'f', '2014-8-1', '2018-2-1'))
conn.commit()

# 데이터 실행후 반드시 commit 해줄것
```

# DB:NULL -> python:None

```python
# 2. 데이터를 list 상태로 한번에 넣을때 many
cur.executemany(
    """INSERT INTO pet (name, owner, species, sex, birth, death)
      VALUES(%s, %s, %s, %s, %s, %s)""", #리스트로 넣어줌
    [('대한','민주', 'cat', 'f', '2001-02-04', None), #DB:NULL -> python:None
    ('민국','자유', 'cat', 'f', '2010-03-17', None),
    ('만세', '민주', 'cat', 'f', '2015-08-27', None),
    ('하자', '민주', 'cat', 'm', '2010-05-13', None),
    ('성공', '자유', 'cat', 'm', '2017-08-31', '2018-04-29')
    ])
conn.commit



print(cur.rowcount) #에러가 난건지 다 들어간건지 확인하는 방법 => rowcount
cur.close()
conn.close
```

# 데이터 가져오기

```python
conn = pymysql.connect(host='192.168.1.248',
                      user='root', port=3306, password='123', db='acornschool',
                      charset='utf8mb4', use_unicode=True, cursorclass=pymysql.cursors.DictCursor)

cur = conn.cursor()






cur.execute( 'select * from pet' ) #RecordSet 생성
print(cur.rowcount) #8
cur.description #설명

cur.fetchone()
cur.fetchone()
cur.fetchone() # 3개를 가져오고
recs = cur.fetchmany(2) # 2개를 가져오면 총 5개
for rec in recs:
    print (rec)
recs = cur.fetchall() # 나머지 3개 가져옴 = 총 8개
for rec in recs:
    print (rec)
    
    
    
    
    
cur.close()
conn.close()
```

# pet_update 저장프로시져 만들어줌

```python
conn = pymysql.connect(host='192.168.1.248',
                      user='root', port=3306, password='123', db='acornschool',
                      charset='utf8mb4', use_unicode=True, cursorclass=pymysql.cursors.DictCursor)

cur = conn.cursor()

data = ("마음2", "마음", 0)
cur.callproc('pet_update', data)
cur.execute('SELECT @_pet_update_2') #이름이 결정 # return값을 받을수 있다
#                               _0, _1, _2  data에서 가져옴??
result = cur.fetchone()
print(result['@_pet_update_2']) #키이데이터 형식
if result['@_pet_update_2'] == 2 :
    print("수정할 이름이 없습니다")
else :
    print("성공적으로 수정되었습니다. ")

conn.commit()

```

# pet_delete 저장프로시져 만들어줌
